```python
# Create a dataframe with the array
data = {"weight":[66,75,84,96,48,71] ,\
        "height":[156,165,186,165,174,167]}
bmi = pd.DataFrame(data)
bmi
```

```python
# Create a new BMI column
bmi["BMI"] = bmi["weight"]/(bmi["height"]/100)**2
bmi
```

```python
# Insert new record
bmi.loc[3] = [78,np.nan,np.nan]
bmi
```

```python
# insert another row with missing values
bmi.loc[5]=[np.nan,67,np.nan]
bmi
```

```python
bmi.info()
```

```python
bmi.isnull().sum()
```

```python
# Replace the missing values with mean value of that column
bmi["weight"] = bmi["weight"].fillna(65)
bmi
```

```python
bmi["height"] = bmi["height"].fillna(165)
bmi 
```

```python
bmi["BMI_new"] = bmi["height"] = bmi["weight"].fillna(165)
bmi 
```

```python
# Drop BMI column
bmi.drop("BMI",axis = 1)
```

```python
bmi.drop("BMI", axis = 1, inplace=True)
bmi
```
